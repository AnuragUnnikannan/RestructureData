```python
import pandas as pd
import numpy as np
import io
```

```python
df = pd.read_csv('input_1.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.rename(columns={'id':'Username'}, inplace=True)
for cols in df.columns:
    li = cols.split(' - ', 1)
    if len(li) < 2:
        li = cols.split('- ', 1)
    newname = '-'.join(li)
    df.rename(columns={cols:newname}, inplace=True)
```

```python
df.info()
```

```python
def unique_elements(li):
    li1 = []
    for i in li:
        if i not in li1:
            li1.append(i)
    return li1
```

```python
li = []
for cols in df.columns:
    li.append(cols)
# print(li)
li1 = []
for i in li:
    li1.append(i.split('-')[0].strip())
# print(li1)
tests = []
for i in li1:
    if "Test" in i:
        tests.append(i)
tests = unique_elements(tests)
print(tests)
```

```python
li = df.columns
newcols = []
for i in tests:
    for j in li:
        if i in j:
            newcols.append(j.split('-',1)[1].strip())
newcols = sorted(unique_elements(newcols))
print(newcols)
```

```python
result = ''
first4 = ["Name", "Username", "Chapter Tag", "Test_Name"]
result += ','.join(first4+newcols)+"\n"
print(','.join(first4+newcols))
for ind in df.index:
    temp = df['Name'][ind] + ','+ str(df['Username'][ind]) + ',' + df["Chapter Tag"][ind]
    
    for test in tests:
        properties = ''
        for col in newcols:
            properties += ',' + df[test+'-'+col][ind]
        print(temp+','+test+properties)
        result += temp+','+test+properties+"\n"
```

```python
df1 = pd.read_csv(io.StringIO(result), na_values='-')
df1
```

```python
df1 = df1.dropna(how='any')
```

```python
df1.to_csv('output1.csv', index=False)
```
